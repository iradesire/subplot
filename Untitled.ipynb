```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Mall_Customers segmentation data.csv")
df['Age'].mean()
df['Age'].median()
df['Age'].quantile([0.19,0.23,0.53])

df['Spending Score (1-100)'].mean()
df['Spending Score (1-100)'].median()
df['Spending Score (1-100)'].quantile([0.39,0.81,0.94])

# x=['Age']
# y=['Spending Score (1-100)']

plt.scatter(df['Age'],df['Spending Score (1-100)'],color='red',s=20)
plt.title('A GRAPH OF AGE VS SPENDING SCORES')
plt.grid()
plt.xlabel('Ages')
plt.ylabel('Spending Score (1-100)')
plt.xticks(rotation=45)
plt.show
```

```python
plt.hist(df['Age'],bins=[20,30,40,50,60,70],color='blue')
plt.title('A GRAPH HOW THE SPENDING DISTRIBUTION')
plt.grid('X')
plt.xlabel('Ages')
plt.ylabel('Spending Score (1-100)')
plt.xticks(rotation=45)
plt.show
```
